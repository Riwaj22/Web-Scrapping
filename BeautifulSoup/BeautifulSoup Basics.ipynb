```python
website = 'https://subslikescript.com/movie/Drain_the_Titanic-4743564'
```

```python
result = requests.get(website)
```

```python
content = result.text
```

```python
soup = BeautifulSoup(content,'lxml')
```

```python
print(soup.prettify())
```

Scrap a single page

```python
box = soup.find('article', class_ = 'main-article')
```

```python
title = box.find('h1').get_text()
```

```python
print(title)
```

```python
transcript = box.find('div', class_ = 'full-script').get_text(
    strip = True,
    separator = ' '
)
```

```python
print(transcript)
```

Exporting data to a txt file

```python
with open('titanic.txt', 'w') as file:
    file.write(transcript)
    
```
